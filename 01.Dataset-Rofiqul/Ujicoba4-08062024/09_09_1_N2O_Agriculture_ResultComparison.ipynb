```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the combined CSV file
file_path = 'hasil/uji4/combined_xgboost_evaluations_with_group_randomsearch.csv'
data = pd.read_csv(file_path)

# Check available test sizes in the dataset
test_sizes = data['Test Size'].unique()
test_sizes = sorted(test_sizes)  # Sort for better visualization

# Define the evaluation metrics
metrics = ['MSE', 'RMSE', 'MAE', 'R2']

# Set up the plotting area
num_metrics = len(metrics)
num_test_sizes = len(test_sizes)
fig, axes = plt.subplots(num_test_sizes, num_metrics, figsize=(15, 15), sharey='row')

# Iterate over each test size and metric
for i, test_size in enumerate(test_sizes):
    data_test_size = data[data['Test Size'] == test_size]
    
    for j, metric in enumerate(metrics):
        ax = axes[i, j]
        
        # Find the best evaluations for each metric
        if metric in data.columns:
            if metric == 'R2':
                best_evaluations = data_test_size.loc[data_test_size.groupby('Group')[metric].idxmax()]
            else:
                best_evaluations = data_test_size.loc[data_test_size.groupby('Group')[metric].idxmin()]
        
        # Plot the bar chart
        ax.bar(best_evaluations['Group'], best_evaluations[metric], color='skyblue')
        
        # Add labels above the bars
        for k, value in enumerate(best_evaluations[metric]):
            ax.text(k, value, f'{value:.2f}', ha='center', va='bottom')
        
        # Set titles and labels
        ax.set_title(f'{metric} (Test Size: {test_size})')
        ax.set_xlabel('Group')
        ax.set_ylabel(metric)
        ax.set_xticks(range(len(best_evaluations['Group'])))
        ax.set_xticklabels(best_evaluations['Group'], rotation=45)

# Adjust layout for better readability
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the combined CSV file
file_path = 'hasil/uji4/combined_xgboost_evaluations_with_group_randomsearch.csv'
data = pd.read_csv(file_path)

# Define the evaluation metrics
metrics = ['MSE', 'RMSE', 'MAE', 'R2']

# Create a dictionary to store the best evaluations for each metric
best_evaluations = {}

for metric in metrics:
    if metric in data.columns:
        # For each metric, find the row with the best value (min for errors, max for R2)
        if metric == 'R2':
            best_evaluations[metric] = data.loc[data.groupby('Group')[metric].idxmax()]
        else:
            best_evaluations[metric] = data.loc[data.groupby('Group')[metric].idxmin()]

# Combine the best evaluations into a single DataFrame
best_evaluations_df = pd.concat(best_evaluations.values()).drop_duplicates().reset_index(drop=True)

# Set the color palette
palette = sns.color_palette("husl", len(best_evaluations_df['Group'].unique()))

# Plotting the best evaluations for each metric
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Helper function to create bar plots
def create_bar_plot(ax, data, metric, palette, title):
    sns.barplot(x='Group', y=metric, data=data, palette=palette, ax=ax)
    ax.set_title(title)
    ax.set_xlabel('Group')
    ax.set_ylabel(metric)
    ax.tick_params(axis='x', rotation=45)

# Plotting MSE
create_bar_plot(axes[0, 0], best_evaluations['MSE'], 'MSE', palette, 'Best MSE by Group')

# Plotting RMSE
create_bar_plot(axes[0, 1], best_evaluations['RMSE'], 'RMSE', palette, 'Best RMSE by Group')

# Plotting MAE
create_bar_plot(axes[1, 0], best_evaluations['MAE'], 'MAE', palette, 'Best MAE by Group')

# Plotting R2
create_bar_plot(axes[1, 1], best_evaluations['R2'], 'R2', palette, 'Best R2 by Group')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
